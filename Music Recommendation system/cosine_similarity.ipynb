```python
# Importing Libraries
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

dataset = pd.read_csv('D:/AI/new project/archive/tcc_ceds_music.csv')

# Selecting the qualities for recommendation
qualities = ['genre', 'artist_name', 'track_name','feelings']
print(qualities)

# Replacing the null values with empty strings
for i in qualities:
    dataset[i] = dataset[i].fillna('')

# Combining all the selected features
total_qualities = dataset['genre'] + ' ' + dataset['artist_name'] + ' ' + dataset['track_name'] + ' ' + ['tpoic']

# Converting the dataset to qualities vectors
vectorizer = TfidfVectorizer()
qualities_vectors = vectorizer.fit_transform(total_qualities)

# Getting the cosine similarity
cos_similarity = cosine_similarity(qualities_vectors)

# Getting input from the user
user_input = input('Enter the name of your favorite song: ')


songs_titles = dataset['track_name'].tolist()
#print(songs_titles)
# Finding similar songs with the user input
similar_songs = difflib.get_close_matches(user_input, songs_titles)
print(similar_songs)

most_similar_song = similar_songs[0]
print(most_similar_song)

# Finding the index of the song and title
song_index = dataset[dataset.track_name == most_similar_song].index[0]
print(song_index)

# Getting a list of similar songs
similarity_list = list(enumerate(cos_similarity[song_index]))
print(similarity_list)

# Sorting the songs based on their similarity score
sorted_songs = sorted(similarity_list, key=lambda x: x[1], reverse=True)
print(sorted_songs)

# Recommended songs
print("========== Top 20 Recommended Songs ===============")
for i in range(0, 20):
    j = sorted_songs[i][0]
    t = dataset.loc[j, 'track_name']
    if i < 9:
        print(i + 1, '-', t)
    else:
        print(i + 1, '-', t)
print("========== Top 20 Recommended Songs ===============\n")
```
